```python
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
import bs4
```

```python
#scrapping the webpages

loader = WebBaseLoader(web_path="https://egeniestores.com/", bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("product-wrapper"))),)

documents = loader.load()
documents
```

```python
#transforming the document containing the data of products

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)
docs
```

```python
#vectorizing the transformed docs

db = Chroma.from_documents(docs, OpenAIEmbeddings())
```
