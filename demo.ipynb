```python
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document
# from langchain_core.vectorstores import InMemoryVectorStore
import bs4
from uuid import uuid4
```

```python
#scrapping the webpages

loader = WebBaseLoader(
    web_path="https://egeniestores.com/", 
    bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("product-wrapper")))
    )

try:
    # Load documents from the web
    documents = loader.load()
except Exception as e:
    print(f"An error occurred while loading documents: {e}")
    documents = []

documents
```

```python
#transforming the document containing the data of products

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)
docs
```

```python
uuids = [str(uuid4()) for _ in range(len(docs))]
```

```python
#vectorizing the transformed docs

REVIEWS_CHROMA_PATH = "./chroma_data"

embeddings = OllamaEmbeddings(
    model="llama3",
)

db = Chroma(
    collection_name="products_collection",
    embedding_function=embeddings,
    persist_directory=REVIEWS_CHROMA_PATH
)



# vectorstore = InMemoryVectorStore()
```

```python
db.add_documents(documents=docs, ids=uuids)
```
