```python
import json
from sentence_transformers import SentenceTransformer
```

```python


# Open and read the JSON file
with open('woocommerce_sample_data.json', 'r') as file:
    data = json.load(file)

# Print the data
print(data)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
product_descriptions = [p['Short description'] for p in data]

# Generate embeddings
embeddings = model.encode(product_descriptions, convert_to_tensor=True)
```

```python
from langchain.vectorstores import FAISS
from langchain_ollama.llms import OllamaLLM
from langchain.chains import RetrievalQA

# Initialize FAISS with embeddings and metadata
faiss_store = FAISS.from_embeddings(embeddings, product_descriptions)

# Use OpenAI or another LLM for the generative component
llm = OllamaLLM(model="llama3.1")

# Create a RAG-based question-answering system
qa_chain = RetrievalQA(llm=llm, retriever=faiss_store.as_retriever())

# Query the chatbot
query = "What are the latest products available?"
response = qa_chain.run(query)
print(response)
```
